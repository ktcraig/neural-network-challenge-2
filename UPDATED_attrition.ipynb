```python
# Create a Standard Scaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(x_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(x_train)
X_test_scaled = X_scaler.transform(x_test)

X_train_scaled
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
encoder = OneHotEncoder(sparse_output=False, drop= None)

# Fit the encoder to the training data
encoder.fit(np.array(y_train[['Department']]))

# Create two new variables by applying the encoder to the training and testing data
y_train_encoded_department = encoder.transform(np.array(y_train[['Department']]))
y_test_encoded_department = encoder.transform(np.array(y_test[['Department']]))

y_train_encoded_department
```

```python
# Create a OneHotEncoder for the Attrition column
encoder = OneHotEncoder(sparse_output=False, drop= None)

# Fit the encoder to the training data
encoder.fit(np.array(y_train[['Attrition']]))

# Create two new variables by applying the encoder to the training and testing data
y_train_encoded_attrition = encoder.transform(np.array(y_train[['Attrition']]))
y_test_encoded_attrition = encoder.transform(np.array(y_test[['Attrition']]))

y_train_encoded_attrition
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_features = len(x_train.columns)
print(num_features)

# Create the input layer with the correct shape
input_layer = layers.Input(shape=(num_features,), name='input_layer')

# Create at least 2 shared layers using Dense layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(128, activation='relu')(shared_layer1)
```

```python
# Create a branch for 'Department' with a hidden layer and an output layer
# Get the number of columns (classes) from the shape of y_train_encoded

#Hidden Layer
department_dense = layers.Dense(64, activation='relu')(shared_layer1)
#Output Layer
department_output = layers.Dense(y_train_encoded_department.shape[1],
                                 activation='sigmoid',
                                 name='department_output')(department_dense)
```

```python
# Create a branch for 'Attrition' with a hidden layer and an output layer
# Get the number of columns (classes) from the shape of y_train_encoded

#Hidden Layer
attrition_dense = layers.Dense(64, activation='relu')(shared_layer1)
#Output Layer
attrition_output = layers.Dense(y_train_encoded_attrition.shape[1],
                                activation='sigmoid',
                                name='attrition_output')(attrition_dense)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy',
                    'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy',
                       'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
"""Colab Assistance Source Code""""
model.fit(
    X_train_scaled[:len(y_train_encoded_department)], # Slice X_train_scaled to match the length of y data
    {'department_output': y_train_encoded_department,
     'attrition_output': y_train_encoded_attrition},
    epochs=10,
    batch_size=32,
)
```

```python
# Evaluate the model with testing data
results = model.evaluate(X_test_scaled[:len(y_test_encoded_department)],
                         {'department_output': y_test_encoded_department,
                          'attrition_output': y_test_encoded_attrition})
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {results[3]}")
print(f"Attrition Accuracy: {results[3]}")
```
